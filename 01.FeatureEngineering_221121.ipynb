### Qu es alta cardinalidad?
Una alta cardinalidad significa que hay muchos valores nicos (como en un campo para nmeros de telfono). La cardinalidad entre tablas es similar, pero se refiere a si una fila de una tabla podra vincularse con ms de una fila de otra tabla.

Los cdigos postales de codificacin nica pueden generar ms de 40.000 caractersticas nuevas. sta no es una solucin prctica. Comprender la estructura de estos cdigos puede permitir que se acumulen a niveles ms altos de agregacin. 

Incluso despus de la agregacin, puede haber demasiadas categoras para que la codificacin one-hot o label sea til. 

Adems, algunas variables categricas no tienen una agregacin fcilmente identificable que pueda reducir el nmero de categoras necesarias. La codificacin de destino o la codificacin de dejar uno fuera pueden manejar estas caractersticas categricas de alta cardinalidad. Si bien algunos blogs de ingeniera de caractersticas discuten estas tcnicas, a menudo no destacan su utilidad en este caso o la impracticabilidad de la codificacin one-hot.
