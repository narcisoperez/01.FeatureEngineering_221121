```python
import pandas as pd
datos = pd.read_csv('/content/drive/MyDrive/ANN/ACARA 9-22.csv',';')

datos.head(5)
```

```python
datos = datos.melt(id_vars=["mtm"], value_name='Precio', var_name='Año').sort_values(by=["mtm", "Año"])
```

```python
datos.tail(5)
```

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:1234@localhost:3306/colab_class', echo=False)
```

```python
try:
  datos.to_sql(name='acara',con=engine,schema = 'colab_class',index=False, if_exists='replace')
  print('Sucessfully written to Database!!!')

except Exception as e:
  print(e)
```

```python
try:
    conexion = pymysql.connect(host='localhost',
                             user='root',
                             password='1234',
                             db='colab_class')
    try:
        with conexion.cursor() as cursor:
            consulta = "SELECT * FROM acara WHERE Año =2018;"
            cursor.execute(consulta)
 
            # Con fetchall traemos todas las filas
            employees = cursor.fetchall()
 
            # Recorrer e imprimir
            for emp in employees:
                print(emp)
    finally:
        conexion.close()
    
except (pymysql.err.OperationalError, pymysql.err.InternalError) as e:
    print("Ocurrió un error al conectar: ", e)
```
